## matplotlib
    ❑A comprehensive library for creating static, animated, and interactive visualizations in Python.
    ❑One of the main Python charting and imaging libraries http://matplotlib.org/


### Matplotlib Basics
    ❑matplotlib.pyplot is a collection of command
    style functions that make matplotlib like MATLAB.
    ❑plot() is a versatile command and will take an
    arbitrary number of arguments.
    ❑show() invoke the created visualizations
    ❑In Notebook, run the magic command
    %matplotlib inline to display the visualizations



```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [3,4,5,7]
plt.plot(x,y)
plt.title('This is a Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

### Line Chart
    ❑Displays the trends of each variable
    ❑The individual variables are
    identified by variations in line
    characteristics (such as thickness or
    style) and/or color

```python
df = pd.read_excel('s3://jrf-ia241-bucket/house_price.xls')
df['built_in'] = pd.to_datetime(df['built_in'],format = "%Y")
df.plot(x = 'built_in', y = 'price')
```

#### Use Line Chart to Show Average Price

```python
avg_price_year = df.groupby('built_in')['price'].mean()
avg_price_year.plot()
```

### Histogram
    ❑The occurrence of data values
    in a statistical distribution

```python
df['price'].hist()
```

### Bar Graph
    ❑Utilize the height or length of the bar to
    represent the data.
    ❑One-dimensional object. The bar widths have no
    value
    ❑Can be complex and compound


```python
avg_price_type = df.groupby('house_type')['price'].mean()
avg_price_type.plot.bar()
```

### Pie Graph
    ❑Components of any total displayed
    as wedges
    ❑The use of a single hue with varying
    saturation is not recommended in
    most cases.

```python
count_per_type = df['house_type'].value_counts()
count_per_type[:]
count_per_type.plot.pie()
```

### Scatter Plot
    ❑Show the covariation of two
    continuous variables
    ❑The triple scatter plot is
    multivariate and depicts the
    relationship of three variables


```python
df.plot.scatter(x='area', y='price')
```
